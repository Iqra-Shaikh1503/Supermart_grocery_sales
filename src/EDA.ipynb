# 2.VISUALIZING THE DATASET

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,6)
```

1.TOTAL SALES BY CATEGORIES

```python
df['Category'] = df['Category'].astype(str).str.strip()
unique_categories = df['Category'].unique()



Sales_by_category = df.groupby("Category")["Sales"].sum()
Sales_by_category = Sales_by_category.reindex(unique_categories)
Sales_by_category.plot(kind="bar", color='yellow')
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

Conclusion:
Sales are well-balanced across all categories. Eggs, Meat & Fish and Snacks lead in total sales, while Oil & Masala ranks lowest, but not by much—indicating consistent performance and strong product diversification.

2.TOTAL SALES BY ORDER DATE 

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
Sales_over_time = df.groupby("Order Date")["Sales"].sum()
Sales_over_time.plot(kind="line", color="green")
plt.title("Sales Over Time")
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

**Conclusion:**
Sales show an overall upward trend from 2015 to 2019, with noticeable spikes in late 2017 and 2018, indicating increased customer demand and business growth over time.


3.CORRELATION HEATMAP

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(12, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

**Conclusion:**
Sales and Profit have a strong positive correlation (**0.61**), meaning higher sales generally lead to higher profit. Discount shows **no significant correlation** with either Sales or Profit, indicating that discounts may not effectively drive sales or impact profit.


4.PROFIT DISTRIBUTION BY CATEGORY

```python
sns.boxplot(x="Category", y="Profit", data=df,palette="Set2")
plt.title("Profit Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.show()
```

Median Profit (central line in each box):

    All categories have a similar median profit, generally between 300–350 units.

    Eggs, Meat & Fish shows a slightly higher median compared to others.

Interquartile Range (IQR – box height):

    The spread (IQR) is relatively similar across all categories, indicating comparable variability in profits.

Outliers (dots above whiskers):

    Every category contains significant outliers, representing unusually high-profit transactions.

    Oil & Masala, Beverages, and Bakery show slightly more frequent and higher outliers, suggesting occasional large profit margins.

Whiskers (range):

    Profits across all categories range from just above 0 to 1000+, indicating some low-profit orders exist, but also high-end profitable ones.



5.TOTAL PROFIT DISTRIBUTION

```python
sns.histplot(df["Profit"], kde=True, color='teal')
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

**Conclusion:**
The profit distribution is **right-skewed**, with most transactions earning a profit between **100–300 units**. As profit increases, the frequency decreases, indicating that **high-profit orders are less common**, while **moderate-profit orders dominate** overall sales.


6.NUMBER OF ORDER BY REGION

```python
sns.countplot(x="Region", data=df, palette="Set3")
plt.title("Number of Order by Region")
plt.xlabel("Region")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```

**Conclusion:**
The **West region** has the highest number of orders, followed by the **East** and **Central** regions. The **South region** shows the lowest order count, indicating a potential opportunity for growth or the need for better outreach in that area.


7.DISCOUNT vs PROFIT SCATTER PLOT

```python
sns.scatterplot(x="Discount", y="Profit", data=df, hue="Region")
plt.title("Discount vs Profit by Region")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.legend(title="Region")
plt.show()
```

**Conclusion:**
There’s **no clear linear relationship** between discount and profit across regions. Profits vary widely at all discount levels, suggesting that offering higher discounts does **not consistently lead to higher or lower profits**. This implies other factors—like product type or volume—play a bigger role in profitability.


8.MONTHLY SALES LINE PLOT

```python
df['Month'] = df['Order Date'].dt.month
monthly_sales = df.groupby('Month')["Sales"].sum()
monthly_sales.plot(kind='line', marker='o', color='purple')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

**Conclusion:**
Sales increase steadily from March and peak in **September, November, and December**, indicating strong performance during **festive and year-end seasons**. Lower sales in the early months suggest slower business at the start of the year.


9.Top 10 cities by Sales

```python
top_cities = df.groupby("City")["Sales"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_cities.values, y=top_cities.index, palette="viridis")
plt.title("Top 10 cities by Sales")
plt.xlabel("Total Sales")
plt.ylabel("cities")
plt.show()
```

**Conclusion:**
**Kanyakumari, Vellore, and Bodi** are the top-performing cities in terms of sales. All top 10 cities show **high and closely competing sales volumes**, indicating strong and balanced demand across these key regions.


10.Pie Chart - Region wise Sales

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution by region")
plt.axis("equal")
plt.show()
```

**Conclusion:**
The **West region** leads in sales (32.1%), followed by **East** (28.4%) and **Central** (23.2%). The **South** region contributes the least (16.3%), while the **North region shows no sales**, indicating either a lack of data or no business activity there.
